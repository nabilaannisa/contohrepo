```python
#create fungsi untuk memasukkan data ke db menggunakan syntax sql Insert Into
def insert_into(tabel,namatabel,conn,cur):
    # creating column list for insertion
    conn = psycopg2.connect(host=host, database=database, user=user, password = password)
    cur = conn.cursor()    
    cols = ",".join([str(i) for i in tabel.columns.tolist()])
    
    # Insert DataFrame recrds one by one.
    for i,row in tabel.iterrows():
#         print(row)
        sql = "INSERT INTO "+namatabel+" (" +cols + ") VALUES (" + "%s,"*(len(row)-1) + "%s)" 
        cur.execute(sql, tuple(row))
    
        # the connection is not autocommitted by default, so we must commit to save our changes
        conn.commit()
```
